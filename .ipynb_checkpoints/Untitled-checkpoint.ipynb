```python
query = """
select Description, Keywords_Present
from {}
where Date_Posted > '2020-06-01'
and Description != 'No Description'
"""

en_stopwords = set(stopwords.words('english'))

def extract_df(job):
    df = pd.read_sql(query.format(job), engine)
    df.drop_duplicates(inplace=True)
    print('Number of rows', len(df))
    return df

def clean_df(df):
    df['Clean'] = [re.sub(r'[!"#$%&()*+,-./:;<=>?@[\]^_`{|}~]', '', text) for text in df['Description']]
    df['Clean'] = [re.sub(r'[0123456789]', '', text) for text in df['Clean']]
    df['Clean'] = [re.sub(r'[^\x00-\x7f]', '', text) for text in df['Clean']]
    # df['Clean'] = df['Clean'].apply(lambda x: " ".join([word for word in x.lower().split() if word not in en_stopwords]))
    df['Clean'] = df['Clean'].apply(lambda x: " ".join([word for word in x.split() if word not in en_stopwords]))
    return df.dropna(subset=['Clean'])

def plot_wordcloud(df):
    all_words = " ".join([str(text) for text in df['Clean']]) 

    wordcloud = WordCloud(width=800, height=500, random_state=42,
                          max_font_size=110).generate(all_words)

    plt.figure(figsize=(10, 7))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
    plt.show()

def lda_modeling(df, bigram=False):
    if bigram:
        count_vect = CountVectorizer(max_df=0.8, min_df=5, stop_words='english', ngram_range=(2,2))
    else:
        count_vect = CountVectorizer(max_df=0.8, min_df=5, stop_words='english')
    doc_term_matrix = count_vect.fit_transform(df['Clean'].values.astype('U'))

    lda = LatentDirichletAllocation(n_components=5, random_state=42)
    lda.fit(doc_term_matrix)

    for i,topic in enumerate(lda.components_):
        print(f'Top 10 words for topic #{i+1}:')
        print([count_vect.get_feature_names()[i] for i in topic.argsort()[-10:]])
        print('\n')
    
    return lda

def get_top_n_words(df, n=20, bigram=False):
    if bigram:
        vec = CountVectorizer(max_df=0.8, min_df=5, ngram_range=(2,2)).fit(df['Clean'])
    else:
        vec = CountVectorizer(max_df=0.8, min_df=5).fit(df['Clean'])
    bag_of_words = vec.transform(df['Clean'])
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    return sorted(words_freq, key = lambda x: x[1], reverse=True)[:n]
```

```python
# Ignore common words
ignore_list = ['Data', 'Business', 'Analyst', 'Management', 'Engineering', 'Engineer', 'Computer', 'Science', \
               'Scientist', 'Learning', 'Finance', 'Degree', 'Job', 'Operations', 'English', 'Manager', \
               'Software', 'Technology', 'Experience', 'Analytics', 'Machine', 'Research', 'Information', 'Service', \
               'Services', 'Product', 'System', 'Singapore', 'Global', 'Financial', 'Team', 'Project', 'Design', \
               'Asia']
min_thres = 0.01
head_count = 10

def get_word_count(df, col_name):
    df_freq = pd.DataFrame(columns=['Word', col_name + ' Count']).set_index('Word')
    word_freq = pd.Series(' '.join(df['Clean']).split()).value_counts()
    word_freq = word_freq[word_freq > min_thres * len(df)]
    for word, count in word_freq.items():
        if word in ignore_list or word.lower() in en_stopwords:
            continue
        if word[0].isupper():
            df_freq.loc[word] = count
            
    return df_freq

def get_bigram_count(df, col_name):
    df_freq = pd.DataFrame(columns=['Word', col_name + ' Count']).set_index('Word')  
    vec = CountVectorizer(max_df=0.8, min_df=min_thres, lowercase=False, ngram_range=(2,2)).fit(df['Clean'])
    bag_of_words = vec.transform(df['Clean'])
    sum_words = bag_of_words.sum(axis=0)
    for word, idx in vec.vocabulary_.items():
        if word.islower():
            continue
        df_freq.loc[word] = sum_words[0, idx]
    return df_freq.sort_values(by=[col_name + ' Count'], ascending=False)
```

```python
df_freq = pd.merge(df_freq_da, df_freq_de, how='outer', on=['Word'])
df_freq = pd.merge(df_freq, df_freq_ds, how='outer', on=['Word'])
df_freq = pd.merge(df_freq, df_freq_mle, how='outer', on=['Word'])
df_freq.fillna(0, inplace=True)
df_freq['Total Count'] = df_freq.loc[:, :].sum(1)
df_freq.sort_values(by=['Total Count'], ascending=False, inplace=True)
print(len(df_freq))
df_freq.head()
```
